```python
import numpy as np #Used for numerical operations and array manipulation.
import pandas as pd #Used for data manipulation 
import random #Generates random numbers.
import tensorflow as tf # building and training deep learning models.
import matplotlib.pyplot as plt #plotting and visualizing data.
from sklearn.metrics import accuracy_score #calculates the accuracy of the model.
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Conv2D, Dense, MaxPooling2D
from tensorflow.keras.optimizers import SGD #ptimizer for training the model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist #dataset of handwritten digits (0-9) 
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data() #x_train (images) and y_train (labels).
```

```python
print(x_train.shape) #60000 images in training set and 28 pixel height and width
```

```python
x_train[0].min(), x_train[0].max() #0-black, 255-white
```

```python
def plot_digit(image, digit, plt, i):
    plt.subplot(4,5,i+1)
    plt.imshow(image, cmap=plt.get_cmap('gray'))
    plt.title(f"Digit:{digit}")
    plt.xticks([])
    plt.yticks([])
plt.figure(figsize=(16,10))
for i in range(20):
    plot_digit(x_train[i], y_train[i], plt, i)
```

```python
from tensorflow.keras.layers import Input
```

```python
model= Sequential([
    Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(100, activation="relu"),
    Dense(10, activation="softmax")
])
```

```python
optimizer= SGD(learning_rate=0.01, momentum=0.9)
model.compile(
    optimizer= optimizer,
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
model.summary()
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

Testing the Model

```python
plt.figure(figsize=(16,10))
for i in range(20):
    image= random.choice(x_test)
    digit= np.argmax(model.predict(image.reshape((1,28,28,1)))[0], axis=-1)
    plot_digit(image, digit, plt, i)
plt.show()
```

Testing

```python
n=random.randint(0,9999)
plt.imshow(x_test[n])
plt.show()
```

```python
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32')  # Reshape and ensure correct dtype
x_test = x_test / 255.0  # Normalize to [0, 1]
```

```python
predicted_value=model.predict(x_test)
print("Handwritten number in the image is= %d" %np.argmax(predicted_value[n]))
```

```python
score= model.evaluate(x_test,y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:',score[1])
```
